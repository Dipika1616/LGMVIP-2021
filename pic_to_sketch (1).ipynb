```python
#Here i imported all the required library...
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
```

```python
# so at 1st we have to read the original image....
img = io.imread('/content/download.png')
cv2_imshow(img)
```

```python
#now we have to convert the image to greyscale 
grey=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(grey)
```

```python
# here we are doing the inverting process of the  grey image,
inv = cv.bitwise_not(grey)
cv2_imshow(inv)
```

```python
#by using gaussian blur function we are bluring the inverted image to get the outline
blur_img=cv.GaussianBlur(inv ,(21,21), 0)
cv2_imshow(blur_img)
```

```python
#now we are inverting the blured image so the noises are removed
inv_blur = cv.bitwise_not(blur_img)
cv2_imshow(inv_blur)
```

```python
#here we are combining the grey image and inverted blur image to get the exact sketch image of the original one.
#and the sketch image is stored in this file.
sketch = cv.divide(grey , inv_blur ,scale=255.0)
cv2_imshow(sketch)
cv.imwrite("sketch.jpg" , sketch)
```
